```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt = pd.read_csv("/content/sample_data/goalscorers.csv")

dt_brasil_argentina = dt[ (dt["team"] == "Brazil") &
  (dt["home_team"] == "Argentina") &
  (dt["date"] >= "2004") ]

dt_agrupado = dt_brasil_argentina.groupby("date")["scorer"].agg(', '.join).reset_index()

plt.figure(figsize=(10, 5))
plt.bar(dt_agrupado["date"], dt_agrupado["scorer"], color="green")
plt.xlabel("Ano")
plt.ylabel("Marcadores")
plt.title("Gols do Brasil na Argentina nos Ãºltimos 10 anos")
plt.xticks(dt_agrupado["date"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt = pd.read_csv("/content/sample_data/goalscorers.csv")

neymar_gols = dt[(dt['scorer'] == "Neymar")]

dt_agrupado = neymar_gols.groupby("date")["scorer"].count().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x="date", y="scorer", data=dt_agrupado, marker="o", linestyle="-", color="green")

plt.xlabel("Data")
plt.ylabel("Quantidade de Gols")
plt.title("Gols do Neymar ao longo dos anos")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt = pd.read_csv("/content/sample_data/goalscorers.csv")

neymar_gols = dt[dt['scorer'] == "Neymar"]

dt_agrupado = neymar_gols.groupby("date")["scorer"].count().reset_index()

dt_agrupado.rename(columns={"scorer": "gols"}, inplace=True)

plt.figure(figsize=(10, 5))
sns.scatterplot(x="date", y="gols", data=dt_agrupado, color="green", s=100)

plt.xlabel("Data")
plt.ylabel("Quantidade de Gols")
plt.title("Gols de Neymar ao longo dos anos")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
